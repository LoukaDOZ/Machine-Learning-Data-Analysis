```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

ds = pd.read_csv("data.csv")
display(ds)
```

```python
# Recherche de valeurs nulles
for c in ds.columns:
    null = pd.isnull(ds[c])
    ds[null]
```

```python
# Valeurs uniques pour rechercher des valeurs aberrantes
for c in ds.columns:
    print(c, " : ", np.sort(pd.unique(ds[c])))
    
# -> Que des variables quantitatives
# Variables discrètes : blue, clock_speed, dual_sim, fc, four_g, int_memory, m_dep, mobile_wt, n_cores, 
#                       pc, sc_h, sc_w, talk_time, three_g, touch_screen, wifi, price_range
# Variables continues : battery_power, px_height, px_width, ram

ds_disc = ds[['blue', 'clock_speed', 'dual_sim', 'fc', 'four_g', 'int_memory', 'm_dep', 'mobile_wt', 'n_cores', 'pc', 'sc_h', 'sc_w', 'talk_time', 'three_g', 'touch_screen', 'wifi', 'price_range']]
ds_cont = ds[['battery_power', 'px_height', 'px_width', 'ram']]
```

```python
ds.describe(include='all')
```

```python
corr = ds.corr().sort_values('price_range', ascending=False)['price_range'].drop('price_range')
corr
```

```python
for c in corr.index:
    plt.scatter(ds['price_range'], ds[c], s=1)
    m, b = np.polyfit(ds['price_range'], ds[c], 1)
    plt.plot(ds['price_range'], m*ds['price_range'] + b, color='red')
    plt.title(c + "   corr = " + str(corr[c]))
    plt.show()
```
